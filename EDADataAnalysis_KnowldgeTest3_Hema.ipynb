```python
expensive_by_location = housing_df.groupby('ocean_proximity')['median_house_value'].mean().sort_values(ascending=False)
print(expensive_by_location)
```

```python
#Find expensive houses
median_price = housing_df['median_house_value'].median()

housing_df['price_category'] = housing_df['median_house_value'].apply(
    lambda price: 'Expensive' if price > median_price else 'Cheap'
)

print(housing_df[['median_house_value', 'price_category']].head())
```

```python
###Average houe price by location using barchart
import matplotlib.pyplot as plt

# Group by ocean proximity and calculate average house value
avg_price_by_location = housing_df.groupby('ocean_proximity')['median_house_value'].mean().sort_values()

# Create a figure of appropriate size
plt.figure(figsize=(10, 6))  

# Plot the vertical bar chart
plt.bar(avg_price_by_location.index, avg_price_by_location.values, color='pink')

# Add labels and title
plt.xlabel('Ocean Proximity')
plt.ylabel('Average House Value')
plt.title('Average House Price by Location')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

##Using scatterplots,using lambdas,dictionary which saves expensive and cheap values.
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have the DataFrame housing_df

# 1. Label each house as 'Cheap' or 'Expensive' based on median price
median_price = housing_df['median_house_value'].median()

housing_df['price_category'] = housing_df['median_house_value'].apply(
    lambda price: 'Expensive' if price > median_price else 'Cheap'
)

# 2. Separate data into cheap and expensive using keys (labels)
category_groups = {
    key: housing_df[housing_df['price_category'] == key]
    for key in housing_df['price_category'].unique()
}

# 3. Plot scatter plots for each category
plt.figure(figsize=(10, 6))

# Use longitude and latitude for plotting locations
colors = {'Cheap': 'green', 'Expensive': 'red'}

for category, subset in category_groups.items():
    plt.scatter(
        subset['longitude'],
        subset['latitude'],
        
        alpha=0.5,
        label=category,
        c=colors[category]
    )

# 4. Add labels, legend, and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Cheap vs Expensive Houses, ')
plt.legend()
plt.tight_layout()
plt.show()
```
