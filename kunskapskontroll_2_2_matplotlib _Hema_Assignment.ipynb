### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4, 100)
y = x**2


figure, axes = plt.subplots( layout='constrained')

axes.plot(x, y)
axes.set_title("Plot of y = xÂ²")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
np.random.seed(15)

# Scatter plot data
random_data_x = np.random.randn(10)
random_data_y = np.random.randn(10)

# Bar plot data
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create figure and 2 subplots
fig, axs = plt.subplots(1, 2, layout='constrained')

# Subplot 1: Scatter Plot
axs[0].scatter(random_data_x, random_data_y, alpha=0.5, color='purple')
axs[0].set_title("Random Scatter Plot")
axs[0].set_xlabel("X values")
axs[0].set_ylabel("Y values")
axs[0].grid(True)

# Subplot 2: Bar Plot
axs[1].bar(names, values, color='orange')
axs[1].set_title("Fruit Count")
axs[1].set_xlabel("Fruit")
axs[1].set_ylabel("Count")
axs[1].grid(axis='y')

plt.show()
```
